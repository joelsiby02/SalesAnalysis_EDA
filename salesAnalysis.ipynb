#  Merging 12 months of data into 1 dataSet

```python
import glob

# get all the csv files in one directory
csv_files = glob.glob('*.csv')

# create an empty list to store dataframes
df_list = []


# loop through each CSV file and append to the list
'''
 for filename in csv_files:
 print(csv_files) <- loop this first to see the files you are going to loop through
'''   

for filename in csv_files:
    df = pd.read_csv(filename)
    df_list.append(df)

    
# concatenate all dataframes in the list into a single dataframe
merged_df = pd.concat(df_list, ignore_index=True)


# save the merged dataframe to a new CSV file
merged_df.to_csv('merged_data.csv', index=False)
```

# 3/ What time should we display ad's to maximize likelihood of customer's buying product

```python
new_df.head(2)
```

```python
# Convert Order Date column to datetime format
#to_datetime is a smart tool to reconstruct your date format, use everywhere
new_df['Order Date'] = pd.to_datetime(new_df['Order Date'])

# Extract hour of the day and day of the week and make a new col out to display it
new_df['Hour'] = new_df['Order Date'].dt.hour  #display hour
new_df['Day of Week'] = new_df['Order Date'].dt.dayofweek  #display dayofwwek
new_df['Minute'] = new_df["Order Date"].dt.minute  #display min
```

```python
new_df
```

```python
# Group by hour and day of the week and calculate total sales
hourly_sales = new_df.groupby(['Hour', 'Day of Week'])['Sales'].sum().reset_index()
hourly_sales.tail()
```

```python
# Create a bar chart of total sales by hour and day of the week
'''
plt.bar(hourly_sales.index, hourly_sales['Sales'])
plt.xticks(hourly_sales.index, hourly_sales['Hour'].astype(str) + '-' + hourly_sales['Day of Week'].astype(str))
plt.xlabel('Hour-Day of Week')
plt.ylabel('Total Sales')
plt.title('Total Sales by Hour and Day of Week')
plt.show()
'''
#Couldn't understand from bar chart
```

```python
# Create a line chart of total sales by hour for each day of the week
sns.set_style('whitegrid') #darkgrid ot whitegrid can be used
plt.figure(figsize=(10, 6))
for day in range(7):
    daily_sales = hourly_sales[hourly_sales['Day of Week'] == day]
    plt.plot(daily_sales['Hour'], daily_sales['Sales'], label=f'Day {day}')
plt.xlabel('Hour of Day')
plt.ylabel('Total Sales')
plt.title('Total Sales by Hour of Day')
plt.xticks(range(24))
plt.legend()
plt.show()
```

#### 4/ What products are mostly sold together?

```python
new_df.head(1)
```

```python
# Arranging duplicate rows together

df = new_df[new_df['Order ID'].duplicated(keep = False)]
df.head(10)
```

```python
# making an other col to display the duplicate items on a single col

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
df.head()
```

```python
# There's a problem above it grouped them twice, therefore lets drop out the duplicate rows

df = df[['Order ID', 'Grouped']].drop_duplicates() #using drop_duplicate fn
df.head()
```

```python
# Surfing to see what items are been purchased together
from itertools import combinations
from collections import Counter

count = Counter()
for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
    print(key, value)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()
for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 3)))
    
for key, value in count.most_common(10):
    print(key, value)
```
